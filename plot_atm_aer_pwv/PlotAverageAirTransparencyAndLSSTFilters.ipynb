# INTERPOLATION

```python
WLMIN=200.
WLMAX=1100.
WL=np.arange(WLMIN,WLMAX,1.)
```

## Interpolate atmosphere

```python
atm_interp = np.interp(WL, wl_atm_mean, trans_atm_mean)
```

## Interpolate filters

```python
U_interp = np.interp(WL, wl,U,)
G_interp = np.interp(WL, wl,G,)
R_interp = np.interp(WL, wl,R,)
I_interp = np.interp(WL, wl,I,)
Z_interp = np.interp(WL, wl,Z,)
Y_interp = np.interp(WL, wl,Y4,)
```

```python
fU=U_interp*atm_interp
fG=G_interp*atm_interp
fR=R_interp*atm_interp
fI=I_interp*atm_interp
fZ=Z_interp*atm_interp
fY=Y_interp*atm_interp
```

```python
fig, ax1 = plt.subplots()
ax1.plot(WL,fU,'b-', lw=2,label="_nolegend_")
ax1.plot(WL,fG,'g-',lw=2,label="_nolegend_")
ax1.plot(WL,fR,'r-',lw=2,label="_nolegend_")
ax1.plot(WL,fI,'y-',lw=2,label="_nolegend_")
ax1.plot(WL,fZ,color='grey',ls='-',lw=2,label="_nolegend_")
ax1.plot(WL,fY,'k-',lw=2,label="_nolegend_")
ax1.set_title("Transmission vs wavelength at LSST site (airmass=1.2)",fontweight='bold')
ax1.set_xlabel("wavelength (nm)",fontweight='bold')
ax1.set_ylabel("Atmosphere Transmission",fontweight='bold')
#plt.legend(loc=2)
plt.ylim(0.,0.8)
plt.xlim(200.,1100)
plt.text(370, 0.33, r'$U$',
         {'color': 'blue', 'fontsize': 24, 'ha': 'center', 'va': 'center',
          'bbox': dict(boxstyle="round", fc="white", ec="black", pad=0.2)})
plt.text(470, 0.5, r'$G$',
         {'color': 'green', 'fontsize': 24, 'ha': 'center', 'va': 'center',
          'bbox': dict(boxstyle="round", fc="white", ec="black", pad=0.2)})
plt.text(615, 0.65, r'$R$',
         {'color': 'red', 'fontsize': 24, 'ha': 'center', 'va': 'center',
          'bbox': dict(boxstyle="round", fc="white", ec="black", pad=0.2)})
plt.text(750, 0.75, r'$I$',
         {'color': 'yellow', 'fontsize': 24, 'ha': 'center', 'va': 'center',
          'bbox': dict(boxstyle="round", fc="white", ec="black", pad=0.2)})
plt.text(870, 0.75, r'$Z$',
         {'color': 'grey', 'fontsize': 24, 'ha': 'center', 'va': 'center',
          'bbox': dict(boxstyle="round", fc="white", ec="black", pad=0.2)})
plt.text(970, 0.45, r'$Y$',
         {'color': 'black', 'fontsize': 24, 'ha': 'center', 'va': 'center',
          'bbox': dict(boxstyle="round", fc="white", ec="black", pad=0.2)})

#ax2 = ax1.twinx()
#ax2.set_ylabel("LSST throughput",fontweight='bold')
plt.savefig("AtmTransparency_LSST_ALL_AVER_AER_PWV.jpg")
```
